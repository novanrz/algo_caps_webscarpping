```python
online['time_update'] = pd.to_datetime(online['time_update'])
online.time_update = online.time_update.dt.month_name()
online.head()
```

```python
## Jawaban no.3 
online2 = online[online['sub_category'] == 'detergent']
pd.crosstab(index=online2['time_update'], 
            columns=online2['scale'],
            values = online2['price_original'],
            aggfunc = 'mean').round(2)
#800gr = Agustus, 1,8kg = Agustus
```

```python
techcrunch = pd.read_csv("data_input/techcrunch.csv")
techcrunch.head()
```

```python
##Jawaban no 4
techcrunch2 = pd.crosstab(index=techcrunch['category'], 
            columns='Rata-rata',
            values = techcrunch['raisedAmt'],
            aggfunc = 'median').round(2)

techcrunch2.sort_values('Rata-rata', ascending= False)
#biotech
```

```python
##Jawaban no 5
techcrunch3 = techcrunch[techcrunch['company'] == 'Friendster']


techcrunch3['fundedDate'] = pd.to_datetime(techcrunch3['fundedDate'])
techcrunch3.fundedDate = techcrunch3.fundedDate.dt.to_period('M')


techcrunch3 = pd.crosstab(index=techcrunch3['fundedDate'], 
            columns='Total',
            values = techcrunch3['raisedAmt'],
            aggfunc = 'sum').round(2)
techcrunch3.sort_values('Total', ascending= False)

#Agustus 2008
```

```python
##Jawaban no 6
techcrunch2 = techcrunch[techcrunch['city'] == 'San Francisco']
techcrunch2 = pd.crosstab(index=techcrunch2['company'], 
            columns='Total',
            values = techcrunch2['raisedAmt'],
            aggfunc = 'sum').round(2)
techcrunch2.sort_values('Total', ascending= False).head()


#Facebook
```
