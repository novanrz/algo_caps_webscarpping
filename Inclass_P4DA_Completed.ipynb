### Categorical and Numerical Variables
Tipe data categorical adalah tipe data yang memiliki karakteristik dimana nilainya selalu berulang pada sebuah kolom di dataframe. Ketika kita bekerja dengan variabel kategorik, ada 2 alasan mengapa kita harus menggunakan tipe data categorical pada pandas. Alasan pertama dengan mempertimbangkan "bussiness prespective". Hal ini dapat menginformasikan dan memandu seorang Analyst, pada pertanyaan seperti metode statistik atau tipe plot mana yang digunakan untuk mengolah data. 

Alasan kedua adalah dengan melihat dari sisi technicalnya. Ketika kita bekerja dengan tipe data categorical pada pandas, hal ini akan jauh menghemat memori dan tentunya akan menambah kecepan komputasional.

Berikut karakteristik Categorical variabel yang dikutip dari official page:
> Categoricals are a pandas data type corresponding to categorical variables in statistics. A categorical variable takes on a limited, and usually fixed, number of possible values (categories; levels in R). Examples are gender, social class, blood type, country affiliation or rating via Likert scales.

Untuk mengubah tipe data ke categorical pada pandas, Anda dapat melakukannya dengan method `astype()` berikut:

**Formula** 
```
df['column_name'] = df['column_name'].astype('new_data_types')
```
**Contoh**
```
employees['marital_status'] = employees['marital_status'].astype('category')
```

Perhatikan kembali dataframe `employees`. Adakah kolom yang tidak tersimpan dengan tipe data yang benar? Ubahlah ke tipe data yang tepat dengan menggunakan method `astype()`!

````python
# code here

employees['division'] = employees['division'].astype('category')
employees.dtypes
````

- Categorical -> category
- Numerical -> integer (int64), float(float64)

````python
employees[['division','name']] = employees[['division','name']].astype('category')
````

- **`select_dtypes()`**

Method `select_dtypes()` digunakan untuk memilih/memfilter kolom-kolom tertentu sesuai dengan nilai parameter yang ada di dalam method `select_dtypes()`. ## Conditional Subsetting

Selain menggunakan operator dan method yang telah dibahas sebelumnya seperti `loc` dan `iloc`, kita dapat melakukan subsetting berdasarkan kondisi tertentu. Misal pada dataframe `rice` yang telah kita buat sebelumnya, kita ingin mengambil beberapa data dengan kondisi sebagai berikut:
- Kita akan mengambil data yang transaksinya berasal dari supermarket -> `.format=='supermarket'`
- Kita akan mengambil transaksi produk yang harganya lebih dari 10000 -> `.unit_price>10000`
- Kita hanya akan mengambil semua transaksi yang pembelian barangnya tidak sama dengan 0 -> `.quantity!=0`, dll

````python
rice = pd.read_csv("data_input/rice.csv", index_col=1)
rice = rice.drop('Unnamed: 0', axis=1)
rice.head(3)
````

Syntax penulisan untuk conditional subsetting adalah:

**```

```**

````python
rice[rice.['format pasar'] == 'supermarket'].head()
````

````python
rice.discount = rice.discount.astype('object')
rice.quantity = rice.quantity.astype('object')
````

Kita juga dapat menggunakan `&` dan `|` untuk melakukan subsetting lebih dari 1 kondisi. Misalnya kita ingin melihat data penjualan dari seorang pegawai yang jumlahnya lebih dari 5000, maka kita dapat menggunakan syntax:
```
sales[(sales.salesperson == 'Moana') | (sales.amount > 5000)]
```
Untuk subsetting dengan kondisi lebih dari 1, setiap kondisi diletakkan di dalam tanda kurung `()` atau bisa ditulis dengan syntax berikut:
```
df[(df['column_name']=='condition 1') operator (df['column_name']=='condition 2')]
```

````python
#code here
#untuk 2 kondisi 

rice[(rice['format']=='supermarket')&(rice['unit_price']>20000)].shape[1]
````
